```python
#Задача 1

class Vertex:
    
    def __init__(self, n):
        self.n = n
        self.edges = []

    def neighbours(self, k):
        result = {}
        Vertex.__all_distances(self, result)
        return [x for x, p in result.items() if min(p) == k]
            
    def __all_distances(vertex, result, p = 0, trace = set()):
        if not vertex in result:
            result[vertex] = set()
        
        result[vertex].add(p)
        for v in vertex.edges:
            if not v in trace:
                Vertex.__all_distances(v, result, p + 1, trace.union(set([vertex])))
    
    def __repr__(self):
        return self.n

a = Vertex('a')
b = Vertex('b')
c = Vertex('c')
d = Vertex('d')
e = Vertex('e')
f = Vertex('f')
a.edges.append(b)
b.edges.append(c)
c.edges.append(a)
a.edges.append(d)
b.edges.append(d)
c.edges.append(d)
d.edges.append(a)
d.edges.append(b)
d.edges.append(c)
a.edges.append(e)
d.edges.append(e)
b.edges.append(f)

#
# e --- d
# |     | \
# |     |  \
# a ---b----c
#     |
#    f

s = a
k = 1
print(s.neighbours(k))
```

```python
#Задача 2

from collections import deque
from bisect import bisect_left

def two_sum_n2(nums, target):
    nums = list(nums)
    r = []
    for i in range(0, len(nums)):
        if nums[i] == None:
                continue
        for j in range(i + 1, len(nums)):
            if nums[j] == None:
                continue
                
            if nums[i] + nums[j] == target:
                r.append((i, j))
                nums[j] = None
                break
    return r

#Если просто заменить словарь на SortedDict с log(N) операциями то получится настоящий two_sum_nlogn, но т.к этот словарь не встроен в питон я сделал так, тут получилось почти o(nlog(n))
#Может не совсем понял условие задачи, но адекватного способа сделать настоящий o(nlog(n)) через бинарный поиск я придумать не могу.
def two_sum_nlogn(nums, target):
    nums = list(zip(nums, range(0, len(nums))))
    nums.sort(key = lambda x: x[0])
    r = []
    
    for i in range(0, len(nums)):
        if nums[i][1] == -1:
            continue        
        m = target - nums[i][0]
        
        pos = bisect_left(nums, m, lo = i + 1, key = lambda x: x[0])
        while pos != len(nums) and nums[pos][1] == -1:
            pos = bisect_left(nums, m, lo = pos + 1, key = lambda x: x[0])
        
        if pos != len(nums) and nums[pos][0] == m:
            a = nums[i][1]
            b = nums[pos][1]
            nums[pos] = (nums[pos][0], -1)
            r.append((min(a, b), max(a, b)))        
    return r

def two_sum_n(nums, target): 
    r = []
    d = {}
    for i in range(0, len(nums)):
        m = target - nums[i] 
        if m in d and len(d[m]) > 0:
            r.append((d[m].popleft(), i))
           
        else:
            if nums[i] in d:
                d[nums[i]].append(i)
            else:
                d[nums[i]] = deque([i])
            
    return r

nums = [2, 7, 7, 5, 5, 2, 2, 4, 4]
print(two_sum_n2(nums, 9))
print(two_sum_nlogn(nums, 9))
print(two_sum_n(nums, 9))
```
